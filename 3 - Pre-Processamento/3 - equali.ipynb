# Equalização do Histograma

O objetivo da equalização do histograma é justamente modificar a tonalida dos pixels da imagem, com intuito de dedistribuir o histograma. Isso resultará em uma imagem mais nítida contendo os pixels com tonalidades de cinza variando de 0 a 255.

Importanto as bibliotecas

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

Lendo a imagem.

A função imread() possui um segundo argumento que especifica como a imagem deve ser lida.

* cv2.IMREAD_COLOR: Le a imagem como colorida.
* cv2.IMREAD_GRAYSCALE: Le a imagem em escala de cinza.
* cv2.IMREAD_UNCHANGED: Le incluindo o canal alfa.

Pode-se simplesmente passar os argumentos usando números inteiros 1, 0 ou -1 respectivamente.

```python
img = cv2.imread("einstein.jpg", 0)
```

Função **EQUALIZEHIST()img**

Entrada: 
    1. Imagem que queremos trabalhar.
    
Saída: 
    1. A imagem com histograma equalizado.

```python
imgEqualizada = cv2.equalizeHist(img)
```

```python
fig = plt.figure(figsize=(20,5))

ax1 = fig.add_subplot(121)
plt.imshow(img, cmap=plt.cm.gray)

ax2 = fig.add_subplot(122)
plt.imshow(imgEqualizada, cmap=plt.cm.gray)
plt.show()
```

```python
fig = plt.figure(figsize=(20,5))

ax1 = fig.add_subplot(121)
plt.hist(img.ravel(), 256, [0, 256])

ax2 = fig.add_subplot(122)
plt.hist(img.ravel(), 256, [0, 256])
plt.show()
```

# Desafio:

* Fica como desafio, fazer os histogramas referentes aos canais de uma imagem colorida (RGB)

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Ler a imagem
img = cv2.imread("einstein.jpg", 1)

# Dividir a imagem em seus três canais (B, G, R)
azul, verde, vermelho = cv2.split(img)

# Equalizar cada canal individualmente
azul_eq = cv2.equalizeHist(azul)
verde_eq = cv2.equalizeHist(verde)
vermelho_eq = cv2.equalizeHist(vermelho)

# Combinar os canais equalizados em uma imagem
imgEqualizada = cv2.merge((azul_eq, verde_eq, vermelho_eq))

# Plotar os histogramas
fig = plt.figure(figsize=(20,5))

ax1 = fig.add_subplot(131)
ax1.hist(azul_eq.ravel(), 256, [0, 256])
ax1.set_title("Histograma do Canal Azul")

ax2 = fig.add_subplot(132)
ax2.hist(verde_eq.ravel(), 256, [0, 256])
ax2.set_title("Histograma do Canal Verde")

ax3 = fig.add_subplot(133)
ax3.hist(vermelho_eq.ravel(), 256, [0, 256])
ax3.set_title("Histograma do Canal Vermelho")

plt.show()
```
